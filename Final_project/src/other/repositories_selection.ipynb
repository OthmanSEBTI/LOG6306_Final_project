```python
from github import Github
import csv

github_session = Github("xxx")
search_results = github_session.search_repositories('stars:>=10',sort="stars", order="desc")

top_100_repos = open('top_100_repos.csv', 'w', newline='', encoding='utf-8') 
writer = csv.writer(top_100_repos)
writer.writerow(['rank','repo','contributors', 'watchers', 'stars'])

for i, repo in enumerate(search_results[:100]):
    
    try:
        writer.writerow([i+1,repo.full_name,repo.get_contributors().totalCount,repo.watchers_count,repo.stargazers_count])
    except Exception:
        continue

top_100_repos.close()
```

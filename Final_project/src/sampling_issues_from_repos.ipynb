```python
from datetime import datetime

date_format = '%Y-%m-%d %H:%M:%S'
date1_str = df['creation_date'][0].replace('T',' ').replace('Z','')
date2_str = df['closing_date'][0].replace('T',' ').replace('Z','')

date1_obj = datetime.strptime(date1_str, date_format)
date2_obj = datetime.strptime(date2_str, date_format)

duration = date2_obj - date1_obj

print('Duration:', duration)
print(type(duration))
```

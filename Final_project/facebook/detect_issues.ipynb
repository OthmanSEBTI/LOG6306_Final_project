```python
import pandas as pd
import re

df = pd.read_csv('/mnt/c/Users/Othman/Desktop/TP-LOG6306/Final_project/facebook/all_commits.csv')
df = df.drop('Unnamed: 0',axis=1)
```

```python

issues = []
number_of_issues = []
messages = df['message'].tolist()
sha = df['sha'].tolist()
pattern = r'#\d+'

for i in range(0,df.shape[0]) :

        if bool(re.search(pattern,messages[i])) == False :
            df.drop(i,inplace=True)
        else :
            issues.append(re.findall(pattern,messages[i]))
            # TO CORRECT ! /_\
            number_of_issues.append(len(issues))
  

df['issues']=issues
df['issues_count']=number_of_issues

df
```

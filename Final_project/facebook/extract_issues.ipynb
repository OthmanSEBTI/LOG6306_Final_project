```python
import pandas as pd
import requests 
import json
import ast
from urllib import request
from urllib.request import Request, urlopen, urlretrieve

df = pd.read_csv('/mnt/c/Users/Othman/Desktop/TP-LOG6306/Final_project/facebook/all_commits_with_issues.csv')
df = df.drop('Unnamed: 0',axis=1)
```

```python


sha = []
message  = []
repo = []
issue_id =[]
issue_state =[]
creation_date = []
closing_date =[]

for i in range(0,df.shape[0]) :

    commit = df.sha [i]
    mess = df.message [i]
    rep = df.repository_full_name [i]
    url = 'https://api.github.com/repos/' + rep +'/issues/'
    issues = ast.literal_eval(df.issues[i])

    for issue in issues:
        
        try :
            url_issue = url + issue.replace('#','')
            request = Request(url_issue, headers={
                                            "authorization": "Bearer xxx"})
            
            data = json.load(urlopen(request))
            sha.append(commit)
            message.append(mess)
            repo.append(rep)
            issue_id.append(issue)
            issue_state.append(data['state'])
            creation_date.append(data['created_at'])
            closing_date.append(data['closed_at'])

        except Exception :
            continue



```

```python

df_issues = pd.DataFrame()

df_issues['sha'] = sha
df_issues['message'] = message
df_issues['repo'] = repo
df_issues ['issue_id'] = issue_id
df_issues['issue_state'] = issue_state
df_issues['creation_date'] = creation_date
df_issues ['closing_date'] = closing_date

df_issues.to_csv('/mnt/c/Users/Othman/Desktop/TP-LOG6306/Final_project/facebook/commits_issues_mapping1.csv')
df_issues[df_issues.issue_state != 'closed']
```
